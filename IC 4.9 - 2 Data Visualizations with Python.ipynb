Create a line chart from a sample of the data, orders hour of day and prices.

```python
# Create a list holding True/False values, to the test np.random.rant() <= 0.7
dev = np.random.rand(len(df_cust_grp_merge)) <= 0.7
```

```python
# Store 70% of the sample in Dataframe big
big = df_cust_grp_merge[dev]
```

```python
# Store 30% of the sample in Dataframe small
small = df_cust_grp_merge[~dev]
```

```python
len(df_cust_grp_merge)
```

```python
len(big) + len(small)
```

```python
df = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df, x = 'orders_day_of_week',y = 'prices')
```

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_ord_dow_price.png'))
```
